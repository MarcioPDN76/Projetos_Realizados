```python
import pandas as pd

vendas = pd.read_csv('01.amazon_sales_dataset (1).csv')
vendas.info()
vendas.columns
vendas
```

```python
analise_vendas = vendas.drop(columns=["order_id", "customer_id", "product_id"])
analise_vendas = analise_vendas.rename(columns={
    "order_date": "Data",
    "ship_date": "Data_Envio",
    "delivery_date": "Data_Entrega",
    "order_status": "Status_Entrega",
    "customer_name": "Nome_Cliente",
    "country": "País",
    "state": "Estado",
    "city": "Cidade",
    "product_name": "Produto",
    "category": "Categoria",
    "sub_category": "Sub_categoria",
    "brand": "Marca_Produto",
    "quantity": "Quantidade",
    "unit_price": "Preço",
    "discount": "Desconto_sobre_Receita",
    "shipping_cost": "Frete",
    "total_sales": "Receita_Menos_Desconto_Mais_Frete",
    "payment_method": "Método_Pagamento"
})
analise_vendas.head()
```

```python
analise_vendas["Data"] = pd.to_datetime(analise_vendas["Data"], format="%Y-%m-%d")

analise_vendas["Data_Envio"] = pd.to_datetime(analise_vendas["Data_Envio"], format="%Y-%m-%d")

analise_vendas["Data_Entrega"] = pd.to_datetime(analise_vendas["Data_Entrega"], format="%Y-%m-%d")
                                                
analise_vendas.sample(20)


analise_vendas.info()
```

```python
coluna_alvo = "Desconto_sobre_Receita"

total_desconto = analise_vendas[coluna_alvo].sum()
desconto_mediano = analise_vendas[coluna_alvo].median()
desconto_medio = analise_vendas[coluna_alvo].mean()
menor_desconto = analise_vendas[coluna_alvo].min()
maior_desconto = analise_vendas[coluna_alvo].max()

print(f'1.Desconto Total de Receitas foi de: {total_desconto:,.2f}')
print(f'2.Desconto Mediano sobre as Receitas foi de: {desconto_mediano:,.2f}')
print(f'3.Desconto Média sobre as Receitas foi de: {desconto_medio:,.2f}')
print(f'4.Menor Desconto Individual sobre as Receitas foi de: {menor_desconto:,.2f}')
print(f'5.Maior Desconto Individual sobre as Receitas foi de: {maior_desconto:,.2f}')
```

O total de descontos concedidos foi de 1.499,68. O desconto médio foi de 0,15 (15%) (o mesmo valor, também, foi encontrado para o desconto mediano). O menor desconto concedido foi de 0,00 e o maior desconto concedido foi de 0,30 (30%).

```python
import numpy as np

coluna_usada = "Desconto_sobre_Receita"

q1 = np.percentile(analise_vendas[coluna_usada], 25)
q2 = np.percentile(analise_vendas[coluna_usada], 50)
q3 = np.percentile(analise_vendas[coluna_usada], 75)

print(f'Primeiro Quartil (Q1 - Menores Descontos): {q1:,.2f}')
print(f'Segundo Quartil (Q2 - Mediana dos Descontos): {q2:,.2f}')
print(f'Terceiro Quartil (Q3 - Maiores Descontos): {q3:,.2f}')
```

Ao ordenar os dados dos Descontos_sobre_Receita vemos que os 25% primeiros descontos montam a $0,07. A mediana desses descontos foi de $0,15 e os 75% mais elevados descontos foi de $0,22.

```python
delta_media_mediana = desconto_medio - q2
distancia_Percentual = desconto_medio/q2

assimetria = analise_vendas[coluna_alvo].skew()

print('--- Relatório Estatístico de Negócio ---')
print(f'A média dos descontos de vendas é {desconto_medio:,.2f}')
print(f'A mediana (valor central) dos descontos é  {q2:,.2f}')
print(f'O Delta (a diferença Financeira) entre elas é {delta_media_mediana:,.2f}')
print('\n--- Análise Comportamental ---\n')
print(f'Distância Relativa entre elas {distancia_Percentual:,.2f}')

if abs (distancia_Percentual -1) < 0.10:
    print(f'INFERÊNCIA: Baixa Dispersão. A média é confiável para representar a coluna.')
elif abs(distancia_Percentual -1) < 0.25:
    print(f'INFERÊNCIA: Dispersão Moderada. Fique atento a extremos, a média começa a mentir.')
else:
    print(f'INFERÊNCIA: Alta Dispersão. A média NÃO é confiável, foque na mediana.')

print(f'Grau de Assimetria (Skewness): {assimetria:,.4f}')
if assimetria > 0.5:
    print(f'INFERÊNCIA (Assimetria Positiva): Nossas vendas têm uma cauda loga para a direita. Os grandes clientes são minoria, todavia eles faturam valores expressivos puxando a venda média para cima.')
elif assimetria < -0.5:
    print(f'INFERÊNCIA (Assimetria Negativa): Nossas vendas têm uma cauda longa para esquerda. O valor médio está sendo puxado para baixo por um grande volume de vendas baratas, talvez pela liquidação.')
else:
    print(f'INFERÊNCIA (Simétrica): Distribuição Equilibrada. Média e mediana estão bem próximas.')
```

Temos uma distribuição dos dados do Desconto_sobre_receita simétrica. Desse modo, a média é bem representativa da distribuição e, por consequência, podemos usá-la, sem problema algum.
